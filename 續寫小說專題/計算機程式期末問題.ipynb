### 組別：讓AI寫小說-紅樓夢
#### 1.	資料在送進去之前有做隨機處理，這個的目的以及好處是什麼？
#### A：透過改變字詞生成的多元度，來找到最合適的文本。例如沒有除的時候temperature為1，不過我們發現temperature為0.5的時候效果最好。

#### 2.	能否讓模型考慮整個句子，而非最後一個字？
#### A：可以，在分割的時候不要分成一個、一個的單詞，而是改分為一個詞、一個詞或更長的文字單位，這樣能讓model不是純粹考慮最後一個字。

#### 4.	有沒有使用其他word embedding ，結果又是如何呢？
#### A：我們沒有使用老師在課堂中提到的one-hot-encoding，因為我們的文本資料量太大，使用one-hot-encoding反而會讓效率變差，所以我們使用了keras中的embedding。

#### 除了keras以外的word embedding方式，我們沒有訓練過其他方法。不過從網路上的資料查詢結果，keras中的embedding層有兩種詞嵌入的方式。一個是我們使用預訓練的方式，通過參數指定預訓練詞向量矩陣，該詞向量矩陣可以是基於任何語言模型，如Word2vec、Glove等。還有一種方式是隨機初始化，Embedding在隨機初始化方式下是一個全連接層，而後面得到的詞表示是全連接層的權重參數。
