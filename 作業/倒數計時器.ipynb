### 倒數計時器
#### 藉由time與datetime套件，實現倒數計時的應用。

```python
import time
from datetime import datetime
tartime=datetime.strptime('2020-06-10-12-29-50','%Y-%m-%d-%H-%M-%S')
while True:
    ntime=datetime.now()
    if ntime<tartime:
        time.sleep(0.001)
        print(tartime-ntime)
    elif ntime>=tartime:
        print('時間到')
        break
```

```python
tartime=datetime.strptime('2019-01-20-11-46-59','%Y-%m-%d-%H-%M-%S')
ntime=datetime.now()
second=(tartime-ntime).seconds
time.sleep(second)
print(1)
```
